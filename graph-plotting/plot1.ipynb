```python
import numpy as np
x = np.linspace(0,10,40)
print (x)
```

```python
c = np.cos(x)
print (c)
```

```python
import matplotlib.pyplot as plt
```

```python
#Only for jupyter notebook
%matplotlib inline
```

```python
plt.plot(x,c)
```

```python
s = np.sin(x)
print(x)
```

```python
plt.figure()
plt.plot(x,c)
plt.plot(x,s)
```

```python
plt.figure()
plt.plot(x, c, label="cos")
plt.plot(x, s, label="sin")
plt.xlabel('x')
plt.ylabel('cos/sin')
plt.legend()
```
